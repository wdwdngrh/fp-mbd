# library

```python
import random
from datetime import datetime, timedelta
```

# set domain

```python
# =========================================================================
# 1. SET DOMAIN KOMPONEN (TERPISAH & MUDAH DIEDIT)
# =========================================================================

# Domain Nama berdasarkan Jenis Kelamin agar Sinkron
nama_depan_l = ["Ahmad", "Budi", "Eko", "Dedi", "Rian", "Hendra", "Fajar", "Rizky", "Andi", "Taufik", "Suryo", "Bambang", "Agus", "Anwar", "Hendro"]
nama_depan_p = ["Siti", "Dewi", "Sri", "Indah", "Putri", "Rina", "Anisa", "Mega", "Fitri", "Lestari", "Wati", "Kartika", "Diana", "Amalia", "Yanti"]
nama_belakang = ["Hidayat", "Santoso", "Prasetyo", "Saputra", "Wijaya", "Kusuma", "Lestari", "Utami", "Nasution", "Siregar", "Purnomo", "Setiawan", "Gunawan", "Sitorus", "Subagyo"]

# Domain Alamat (Jalan dan Kota dipisah agar bisa dikombinasikan)
nama_jalan = ["Jl. Merdeka", "Jl. Sudirman", "Jl. Diponegoro", "Jl. Gajah Mada", "Jl. Pemuda", "Jl. Pahlawan", "Jl. Gatot Subroto", "Jl. Hayam Wuruk", "Jl. Ahmad Yani", "Jl. Asia Afrika"]
nama_kota = ["Bandung", "Jakarta", "Surabaya", "Yogyakarta", "Semarang", "Medan", "Malang", "Makassar", "Bogor", "Solo"]

# Domain Medis & Farmasi
nama_obat_pool = [
    {"nama": "Paracetamol 500mg", "satuan": "Tablet", "harga": 5000},
    {"nama": "Antasida Doen", "satuan": "Tablet", "harga": 4000},
    {"nama": "Amoxicillin 500mg", "satuan": "Tablet", "harga": 12000},
    {"nama": "OBH Combi Sirup", "satuan": "Botol", "harga": 25000},
    {"nama": "Cetirizine 10mg", "satuan": "Tablet", "harga": 8000},
    {"nama": "Ibuprofen 400mg", "satuan": "Tablet", "harga": 7000},
    {"nama": "Ranitidine 150mg", "satuan": "Tablet", "harga": 6000},
    {"nama": "Ambroxol Sirup", "satuan": "Botol", "harga": 18000}
]

domain_keluhan = [
    "Demam tinggi dan sakit kepala.", "Batuk berdahak dan flu berat.",
    "Nyeri lambung dan mual-mual.", "Gatal-gatal di seluruh tubuh.",
    "Pemeriksaan rutin tekanan darah.", "Sesak napas ringan setelah beraktivitas.",
    "Nyeri sendi pada lutut kanan.", "Sakit gigi berdenyut sejak 2 hari."
]

domain_diagnosa = [
    ("Febris / Demam virus", "Pemberian antipiretik dan edukasi istirahat"),
    ("Acute Nasopharyngitis", "Edukasi hidrasi dan pemberian ekspektoran"),
    ("Dispepsia / Maag Akut", "Pemberian antasida dan edukasi pola makan"),
    ("Urtikaria Akut / Alergi", "Pemberian antihistamin"),
    ("Hipertensi Esensial", "Edukasi diet rendah garam dan pemantauan tensi"),
    ("Asma Bronkial Ringan", "Edukasi pembatasan pemicu alergi"),
    ("Osteoarthritis", "Pemberian analgetik topikal/oral"),
    ("Pulpitis / Radang Gigi", "Rujukan ke dokter gigi setelah nyeri reda")
]

domain_aturan = ["3 x 1 tablet setelah makan", "1 x 1 tablet malam hari", "3 x 1 tablet kunyah sebelum makan", "3 x 1 sendok takar sesudah makan"]
```

# insert dml

```python
# =========================================================================
# 2. PROSES GENERASI OTOMATIS BERDASARKAN LIMIT (200 DATA)
# =========================================================================
LIMIT_DATA = 200

lines = [
    "-- =========================================================================",
    f"-- SCRIPT DML AUTOMATION: {LIMIT_DATA} DATA GENERATED DYNAMICALLY VIA PYTHON",
    "-- =========================================================================",
    "BEGIN;\n"
]

# --- A. GENERATE 200 DATA PASIEN ---
lines.append(f"-- 1. Tabel Pasien (Limit: {LIMIT_DATA} Baris)")
for i in range(1, LIMIT_DATA + 1):
    jk = random.choice(["L", "P"])
    nama_depan = random.choice(nama_depan_l) if jk == "L" else random.choice(nama_depan_p)
    nama_lengkap = f"{nama_depan} {random.choice(nama_belakang)}"

    nik = f"327301{random.randint(1000000000, 9999999999)}"
    nomor_rm = f"RM-2026-{str(i).zfill(4)}"
    tgl_lahir = (datetime.now() - timedelta(days=random.randint(5000, 22000))).strftime('%Y-%m-%d')
    alamat = f"{random.choice(nama_jalan)} No. {random.randint(1, 150)}, {random.choice(nama_kota)}"
    telp = f"0812{random.randint(10000000, 99999999)}"

    sql = f"INSERT INTO pasien (nomor_rm, nama_pasien, nik, jenis_kelamin, tanggal_lahir, alamat, no_telepon) VALUES ('{nomor_rm}', '{nama_lengkap}', '{nik}', '{jk}', '{tgl_lahir}', '{alamat}', '{telp}');"
    lines.append(sql)
lines.append("")

# --- B. GENERATE TABEL OBAT ---
lines.append("-- 2. Tabel Obat")
for i, o in enumerate(nama_obat_pool, 1):
    stok = random.randint(50, 300)
    sql = f"INSERT INTO obat (kode_obat, nama_obat, satuan, harga_jual, stok_obat) VALUES ('OBT-{str(i).zfill(3)}', '{o['nama']}', '{o['satuan']}', {o['harga']}.00, {stok});"
    lines.append(sql)
lines.append("")

# --- C. GENERATE 200 DATA KUNJUNGAN ---
lines.append(f"-- 3. Tabel Kunjungan (Limit: {LIMIT_DATA} Baris)")
base_time = datetime.now() - timedelta(days=15) # Simulasi rekam medis 15 hari ke belakang
for i in range(1, LIMIT_DATA + 1):
    base_time += timedelta(minutes=random.randint(20, 90)) # Waktu pendaftaran bertahap
    tgl_kunjungan = base_time.strftime('%Y-%m-%d %H:%M:%S')
    keluhan = random.choice(domain_keluhan)
    status = "Selesai" if i <= (LIMIT_DATA - 3) else random.choice(["Diperiksa", "Menunggu"])

    sql = f"INSERT INTO kunjungan (id_pasien, tanggal_kunjungan, keluhan_utama, status_kunjungan) VALUES ({i}, '{tgl_kunjungan}', '{keluhan}', '{status}');"
    lines.append(sql)
lines.append("")

# --- D. GENERATE DATA PEMERIKSAAN ---
# Hanya generate pemeriksaan bagi kunjungan yang sudah 'Selesai' atau 'Diperiksa' (misal 198 data)
limit_periksa = LIMIT_DATA - 2
lines.append(f"-- 4. Tabel Pemeriksaan (Limit: {limit_periksa} Baris)")
base_time = datetime.now() - timedelta(days=15)
for i in range(1, limit_periksa + 1):
    base_time += timedelta(minutes=random.randint(20, 90))
    tgl_periksa = (base_time + timedelta(minutes=15)).strftime('%Y-%m-%d %H:%M:%S') # Diperiksa 15 mnt setelah daftar
    tensi = f"{random.choice([110, 120, 130, 140])}/{random.choice([70, 80, 85, 90])}"
    bb = round(random.uniform(45.0, 85.0), 1)
    suhu = round(random.uniform(36.2, 39.0), 1)
    diag, tindakan = random.choice(domain_diagnosa)

    sql = f"INSERT INTO pemeriksaan (id_kunjungan, tanggal_periksa, tensi_darah, berat_badan, suhu_tubuh, diagnosa, tindakan_medis) VALUES ({i}, '{tgl_periksa}', '{tensi}', {bb}, {suhu}, '{diag}', '{tindakan}');"
    lines.append(sql)
lines.append("")

# --- E. GENERATE DATA RESEP & DETAIL RESEP ---
# Kita asumsikan dari 198 pemeriksaan, 180 di antaranya menghasilkan resep obat jalan
limit_resep = limit_periksa - 18
lines.append(f"-- 5. Tabel Resep Header (Limit: {limit_resep} Baris)")
for i in range(1, limit_resep + 1):
    sql = f"INSERT INTO resep (id_pemeriksaan) VALUES ({i});"
    lines.append(sql)
lines.append("")

lines.append("-- 6. Tabel Detail Resep (Simulasi Multi-Obat per Resep)")
for i in range(1, limit_resep + 1):
    # Setiap resep akan mendapatkan 1 sampai 2 jenis obat secara acak
    banyak_obat = random.randint(1, 2)
    obat_terpilih = random.sample(range(1, len(nama_obat_pool) + 1), banyak_obat)
    for o_id in obat_terpilih:
        jml = random.randint(5, 15)
        aturan = random.choice(domain_aturan)
        sql = f"INSERT INTO detail_resep (id_resep, id_obat, jumlah, aturan_pakai) VALUES ({i}, {o_id}, {jml}, '{aturan}');"
        lines.append(sql)
lines.append("")

# --- F. GENERATE DATA PEMBAYARAN ---
# Semua kunjungan dicatat nota pembayarannya (200 baris)
lines.append(f"-- 7. Tabel Pembayaran (Limit: {LIMIT_DATA} Baris)")
for i in range(1, LIMIT_DATA + 1):
    if i <= limit_resep:
        # Jika kunjungan mendapat obat jalan
        biaya_layanan = random.choice([50000, 75000])
        biaya_obat = random.randint(20000, 150000)
        status_bayar = "'Lunas'"
        waktu_bayar = f"'{datetime.now().strftime('%Y-%m-%d %H:%M:%S')}'"
    elif i <= limit_periksa:
        # Jika diperiksa tapi tidak tebus obat jalan (misal hanya tindakan di klinik)
        biaya_layanan = 75000
        biaya_obat = 0
        status_bayar = "'Lunas'"
        waktu_bayar = f"'{datetime.now().strftime('%Y-%m-%d %H:%M:%S')}'"
    else:
        # Jika statusnya masih mengantre (belum diperiksa / belum bayar)
        biaya_layanan = 0
        biaya_obat = 0
        status_bayar = "'Belum Lunas'"
        waktu_bayar = "NULL"

    sql = f"INSERT INTO pembayaran (id_kunjungan, total_biaya_layanan, total_biaya_obat, status_pembayaran, waktu_pembayaran) VALUES ({i}, {biaya_layanan}.00, {biaya_obat}.00, {status_bayar}, {waktu_bayar});"
    lines.append(sql)
lines.append("")

lines.append("COMMIT;")
```

# save txt

```python

# =========================================================================
# 3. PENYIMPANAN KE FILE .TXT
# =========================================================================
output_file = "dml_klinik_200_data.txt"
with open(output_file, "w", encoding="utf-8") as f:
    f.write("\n".join(lines))

print(f"🎉 SUKSES BESAR! Berhasil mengombinasikan domain secara acak.")
print(f"File '{output_file}' berisi total ratusan baris skrip SQL DML siap digunakan!")
```
